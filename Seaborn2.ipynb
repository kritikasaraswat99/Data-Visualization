Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# Import Data

```python
df = sns.load_dataset("tips")
```

```python
df.head(10)
```

# Distribution Plots

*   Distplot
*   jointplot
*   pairplot





# **Dist plot**

Dist plot helps us to check the distribution of the columns feature

```python
#univariate plot
#based on the X axes it will show you the density
sns.distplot(df['tip'])
```

```python
#bin is like a bucket which devide the data into diffrent intervals
#kde is kernle density estimation
sns.distplot(df["tip"],kde=False, bins=10)
```

# Joint Plot

 - A join plot allows to study the relationship between 2 numeric variables. the central chart display their correlation, it is usually a scatterplot, a hexbin plot, a 2D histogram or 2D density plot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex') #kind{ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
```

```python
#probability density function
#and draws a regression line that is the best fit line

sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

# pairplot

In pairplot one variable in the same data row is matched with another variable's value, like this: Pair plots are just elborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df, palette='Blues')
```

```python
sns.set(style="dark")
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,hue="smoker")
```


# **Categorical Plots**

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about


*   boxplot
*   countplot
*   bar


# Bar Plot

```python
#represent the variance between the male/female and total bill amount
sns.barplot(y='total_bill', x='sex',data=df, estimator=np.median)
```

# Count Plot

```python
sns.countplot('sex', data=df)
```

```python
sns.countplot('day', data=df)
```

# Box Plot

A box and whisker plot is a graph that shows the spread and centers of the dataset. it is very useful when we want to find the outliers in our dataset.

```python
# A box plot allows you to compare different variables
# The box shows the quartiles of the data. The bar in the middle is the median and
# the box extends 1 standard deviation from the median
# The whiskers extend to all the other data aside from the points that are considered
# to be outliers
# Hue can add another category being sex
# We see men spend way more on Friday versus less than women on Saturday
sns.boxplot(x="day", y="total_bill", hue="smoker",data=df)

plt.legend(loc=0)
```

```python
sns.boxplot('smoker','total_bill', data =df)
```

```python
sns.boxplot(x = 'day',y = 'total_bill', data=df, palette='rainbow')
```

```python
#box plot for all numerical for all the features
#orient“v” | “h” (optional) Orientation of the plot (vertical or horizontal)
sns.boxplot(data=df,orient='h')
```

# **Correlation with Heatmap** -

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimentions or event types, it i very important in Feature Selection.

```python
df.corr()
```

```python
sns.heatmap(df.corr(), cmap='Blues', linecolor='white', linewidth =2)
```
