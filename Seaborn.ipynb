**BOX PLOT**

```python
# A box plot allows you to compare different variables
# The box shows the quartiles of the data. The bar in the middle is the median and
# the box extends 1 standard deviation from the median
# The whiskers extend to all the other data aside from the points that are considered
# to be outliers
# Hue can add another category being sex
# We see men spend way more on Friday versus less than women on Saturday
```

```python
sns.set(style="white", color_codes=True)
a=sns.load_dataset('tips')
sns.boxplot(x="day", y ="total_bill", data =a)
```

```python
sns.set(style="white", color_codes=True)
a=sns.load_dataset('tips')
sns.boxplot(x="day", y ="total_bill", data =a)
sns.despine(offset=10, trim=True)
```

**Correlation with Heatmap** -

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimentions or event types, it i very important in Feature Selection.

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='Blues', linecolor='white', linewidth =1)
```

JoinPlot

 - A join plot allows to study the relationship between 2 numeric variables. the central chart display their correlation, it is usually a scatterplot, a hexbin plot, a 2D histogram or 2D density plot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex') #kind{ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
```

```python
#probability density function
#and draws a regression line that is the best fit line

sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

pairplot

a pairplot is also known as a catterplot, in which one variable in the same data row is matched with another variable's value, like this: Pair plots are just elborations on this, showing all variables paired with all the other variables

**Dist plot**

Dist plot helps us to check the distribution of the columns feature

```python
#univariate plot
#based on the X axes it will show you the density
sns.distplot(df['tip'])
```

```python
#bin is like a bucket which devide the data into diffrent intervals
#kde is nothing but kernle density estimation
sns.distplot(df["tip"],kde=False, bins=10)
```

Box Plot

A box and whisker plot(sometimes called a boxplot) is a graph that presents informaton from a fiv-number summary.

```python
sns.boxplot('smoker','total_bill', data =df)
```

```python
sns.boxplot(x = 'day',y = 'total_bill', data =df, palette='rainbow')
```

```python
#for all the features
sns.boxplot(data=df,orient='v')
```

```python
# A box plot allows you to compare different variables
# The box shows the quartiles of the data. The bar in the middle is the median and
# the box extends 1 standard deviation from the median
# The whiskers extend to all the other data aside from the points that are considered
# to be outliers
# Hue can add another category being sex
# We see men spend way more on Friday versus less than women on Saturday
sns.boxplot(x="day", y="total_bill", hue="smoker",data=df)

plt.legend(loc=0)
```

**Violin Plot**

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
# Violin Plot is a combination of the boxplot and KDE
# While a box plot corresponds to data points, the violin plot uses the KDE estimation
# of the data points
# Split allows you to compare how the categories compare to each other
sns.violinplot(x="day", y="total_bill", data=df,palette='rainbow')
```
