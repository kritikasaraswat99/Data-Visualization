```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1)
random_values = np.random.randn(50)
sns.displot(random_values, kde=True, height=4, aspect=2)
plt.show()
```

```python
penguins_data = sns.load_dataset("penguins")
```

```python
penguins_data.head()
```

Continuous value

```python
sns.displot(data=penguins_data, x="flipper_length_mm", kde=True, height=4, aspect=2) 
plt.show()
```

```python
sns.jointplot(data=penguins_data, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(np.random.randn(20).reshape(4,5)) # random numbers of shape (4,5)
correlation_matrix = df.corr() # shows correlation
correlation_matrix
```

```python
ax_grid = sns.heatmap(correlation_matrix, cbar=0, linewidths=2,vmax=1, vmin=0, square=True, cmap='Blues', annot=True)
plt.show()
```

```python
sns.set_theme(style="darkgrid")
titanic_values = sns.load_dataset("titanic") # seaborn has inbuild varaity of datasets
ax_grid = sns.countplot(x="class", hue = 'who',data=titanic_values)
```

```python
sns.pairplot(data =penguins_data,hue = 'species', diag_kind = 'hist')
```

```python
tips_data = sns.load_dataset("tips")
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips_data, palette="Set3")
```

```python
tips_data
```

```python
sns.scatterplot(data = tips_data, x = 'total_bill', y = "tip", hue = "time")
```

```python
#for confidence interval
ax_grid = sns.pointplot(x="time", y="total_bill", data=tips_data)
```

```python
#blackline represents the error estimation

sns.barplot(data = tips_data, x = 'day', y = "total_bill")
```
