**Line Plots with numpy array **

```python
import numpy as np

linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2

plt.figure()
plt.plot(linear_data,'-o', quadratic_data, '-o')
```

**Histogram**

The bin refers to the range of values divided into a series of intervals.

```python
import matplotlib.pyplot as plt
```

```python
population_ages = [22,33,44,55,66,77,22,44,676,44]
bins=[0,10,20,30,40,50,60]

plt.hist(population_ages, bins, histtype='bar', rwidth=0.8)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Histogram')

plt.show()
```

**Scatter Plot** - in oder to compare two variable to see the correlation

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7,8]
y=[5,2,4,2,1,5,2,7]

plt.scatter(x,y, label='skitscat', color='g')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

**Stack Plot / Area plot**  when you are tracing the changes in two or more related group

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color = 'm', label='Sleeping', linewidth=5)
plt.plot([],[],color='c',label='Eating', linewidth=5)
plt.plot([],[],color='r',label='Working', linewidth=5)
plt.plot([],[],color='k',label='Playing', linewidth=5)

plt.stackplot(days, sleeping, eating,working,playing, colors=['m',"c","r","k"])

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Stack Plot')
plt.legend(loc=4, frameon=False, title='Legend')
plt.show()
```

**`Pie Chart`** - used to show parts of the whole and often a percentage to share

```python
import matplotlib.pyplot as plt

slices = [7,2,2,13] #relevent sizes for each part
activities = ['sleeping','eating','working','playing']
cols = ["c","m","r","b"]

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90, #start angle for the graph, lets you start the line from where you want
        shadow=True,
        explode=(0,0.1,0,0), #to pullout the slce a bit
        autopct='%1.1f%%')

plt.title("Pie Plot")
plt.show()
```
